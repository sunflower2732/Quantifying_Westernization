```python
def calculate_mean(row, df, label):
    # Extract the years list from the row. Row = row from the df_w_years dataset (the one with the questions)
    years_list = row[label]
    years_list = [str(year) for year in years_list]

    df = df[df['Country_Wave'] == row['Country_Wave']]
    df[years_list].fillna(-1)

    values = df[years_list].values.flatten()
    
    values = values[values > 0]
    
    # Calculate and return the mean of those values
    return values.mean() if len(values) > 0 else None
```
